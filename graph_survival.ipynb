```python
import numpy as np
import pickle
from cox_code import * # Code available upon request 
import csv
import matplotlib.pyplot as plt
%matplotlib inline

# time, censor, spayneuter, Gender, color
index_interest = [0, 1, 9, 10, 13]

# Import data
data_array = []
with open('combine_data_process.csv', 'rb') as csvfile:
    outreader = csv.reader(csvfile, delimiter=',')
    outreader.next()  # skip header
    for row in outreader:
        temp = []
        for i in index_interest:
            temp.append(int(row[i]))
        data_array.append(temp)
data = np.array(data_array)
print data.shape
```

```python
# my program found these values
beta = np.array([-0.25374498,  0.07650157,  0.12678004])

time_index = False
time_data = time_index # no timed data
```

```python
surv0, surv1, censor0_val, censor1_val, censor0_time, censor1_time, time_graph = survival_graph(data, time_data, beta, 4)
plt.step(time_graph, surv1, label = "Black")
plt.step(time_graph, surv0, label = "Not Black")
plt.xlabel('Weeks', fontsize = 13)
plt.ylabel('Probability of Adoption or Transfer', fontsize= 12)
plt.title('Survival of Animals of Different Colors', fontsize = 15)
plt.legend(loc="lower left")
plt.show()
```

```python
surv0, surv1, censor0_val, censor1_val, censor0_time, censor1_time, time_graph = survival_graph(data, time_data, beta, 2)
plt.step(time_graph, surv1, label = "Spayed or Neutered")
plt.step(time_graph, surv0, label = "Intact")
plt.xlabel('Weeks', fontsize = 13)
plt.ylabel('Probability of Adoption or Transfer', fontsize= 12)
plt.title('Survival of Spayed/Neutered and Intact Animals', fontsize = 14)
plt.legend(loc="lower left")
plt.show()
```
